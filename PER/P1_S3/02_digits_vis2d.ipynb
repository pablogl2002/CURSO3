# Visualización de digits en 2d

Si los datos vienen representados con muchas características (homogéneas), podemos aplicar una técnica de reducción de la dimensión como PCA y visualizarlos en 2D (o 3D) para hacerse una idea sobre qué clases son potencialmente más difíciles de discriminar.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
```

Lectura de digits y reducción a 2d:

```python
digits = load_digits()
X2d = PCA(n_components=2).fit_transform(digits.data)
```

Gráfico de dispersión:

```python
plt.figure(figsize=(10, 10))
colors = cm.tab10(np.linspace(0, 1, 10))
for digit in digits.target_names:
    plt.scatter(*X2d[digits.target == digit].T, marker=f"${digit}$", s=50, alpha=0.5, color=colors[digit])
```

**Ejercicio:** Por claridad, en la gráfica matricial abajo se muestran separadamente los datos 2d de cada par de clases. A partir de la misma podemos determinar qué pares de clases se discriminan bien en 2d y qué pares no. Si un par de clases se separa bien en 2d, ¿qué cabe esperar en más dimensiones (sin aplicar reducción)?; ¿y si el par de clases no se separa bien en 2d?

```python
fig, axes = plt.subplots(10, 10, figsize=(20, 20))
for i in np.arange(0,10):
    for j in np.arange(0,10):
        ax = axes[i,j]
        ax.set_axis_off()
        ax.set_title('{:d}-{:d}'.format(i,j))

        ax.scatter(*X2d[digits.target == i].T, marker=f"${i}$", s=50, alpha=0.5, color=colors[i])
        ax.scatter(*X2d[digits.target == j].T, marker=f"${j}$", s=50, alpha=0.5, color=colors[j])
```
