# El corpus iris

El corpus iris ha sido ampliamente utilizado para introducir conceptos y métodos básicos de reconocimiento de formas, esto es, aprendizaje automático de clasificadores. Consta de $N=150$ muestras, $50$ por cada una de $C=3$ clases, representadas mediante vectores de $D=4$ características reales homogéneas. Una de las clases es linealmente separable del resto, pero las otras dos no son linealmente separables. Aunque hoy en día se considera un corpus de "juguete", sigue siendo muy útil para introducir conceptos y métodos básicos.

Primero importamos algunas librerías estándar y sklearn:

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

Lectura del corpus iris:

```python
iris = load_iris()
print(dir(iris))
X = iris.data 
y = iris.target
fn = iris.feature_names
cn = iris.target_names
print(iris.DESCR)
```

Convertimos el corpus en un dataframe pandas para facilitar su descripción:

```python
data = pd.DataFrame(data=X, columns=fn)
data['species'] = pd.Series(iris.target_names[y], dtype='category')
# diferentes formas de buscar un elemento en una tabla pandas
print(data.loc[0][0])
print(data.loc[0]["sepal length (cm)"])
data
```

Veamos algunas estadísticas básicas:

```python
data.describe()
```

Comprobamos que tenemos $50$ muestras de cada clase:

```python
data.groupby('species').size()
```

Como tenemos pocas características, es buena idea hacer un gráfico matricial de dispersión:

```python
sns.pairplot(data, hue="species", height = 2, palette = 'inferno');
```
